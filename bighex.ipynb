```python
def plotPoly(verts, fig=None, ax=None, *args, **kwargs):
    if not len(verts): raise ValueError("No vertices. verts: %s" % verts)
    
    if fig is None: fig = plt.figure(figsize=(12, 12))
    if ax is None: ax = fig.add_axes([0, 0, 1, 1])
    
    verts = verts + [verts[0]]
    
    fig = plt.gcf()
    ax = plt.gca()
    
    ax.set_aspect('equal', 'box')
    ax.plot(*zip(*verts), *args, **kwargs)
    
    return fig,ax

def plotPolys(polys, fig=None, ax=None, fill=None, *args, **kwargs):
    if fig is None: fig = plt.figure(figsize=(12, 12))
    if ax is None: ax = fig.add_axes([0, 0, 1, 1])
    
    for verts in polys:
        plotPoly(verts, fig=fig, ax=ax, *args, **kwargs)
        if fill:
            ax.fill(*zip(*verts), fill)
    
    ax.set_aspect('equal', 'box')
    return fig,ax
```
